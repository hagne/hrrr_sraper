```python
fngribhrrr2d = 'ncep_hrrr_2d.2133506000000.grib2'
grbs = pygrib.open(fngribhrrr2d)
```

```python
grbs.rewind()
```

```python

for grb in grbs:
    print(grb)
```

### all keys

```python
for k in grb.keys():
    if k in ['analDate', 'validDate']: # these two key cause errors
        continue
    # print(f'{k}:{grb[k]}')
    print(k, end = ': ')
    print(grb[k])
    
```

### all "relevant" keys

```python
def grib2var_and_keys(grbs):
    grbs.rewind()
    grb_list = []
    for e,grb in enumerate(grbs):
        # break
        # print(e)
        grb_line = {'netcdf variable name': '', 'messagenumber': grb.messagenumber, 'typeOfLevel': grb.typeOfLevel, 'level': grb.level, 
                    # 'unitsECMF': grb.unitsECMF, # units key is more complete and this never seem to give addition information
                    'units': grb.units}
        for k in grb.keys():
            if k == 'modelName': #always unknown
                continue

            elif 'name' in k.lower():
                pass
            else:
                continue
            # print(f'{k}:{grb[k]}')
            # print(k, end = ': ')
            # print(grb[k])
            grb_line[k]=grb[k]

        grb_list.append(grb_line)


    grb_df = pd.DataFrame(grb_list)
    # remove higher hybrid levels
    grb_df = grb_df[~ np.logical_and(grb_df.typeOfLevel == 'hybrid', grb_df.level!=1)]
    
    return grb_df

def write2excel(fnxls, msg):
    writer = pd.ExcelWriter(fnxls, engine='xlsxwriter')
    grb_df.to_excel(writer,index = False)

    worksheet = writer.sheets['Sheet1']

    worksheet.write_comment('A1', msg)

    worksheet.set_column(0, grb_df.shape[1], 20)
    
    writer.save()
    return writer
```

```python
grb_df = grib2var_and_keys(grbs)
```

```python
grb_df
```

```python
msg = f'Extraction of "relevant" keys from {fngribhrrr2d}. First line shows the keys. Each line shows the value of those keys for all the messages (variables/parameters)'
fnxls = 'hrrr_2d_grb_info.xlsx'
write2excel(fnxls, msg)
```
